## í¬ë¡œìŠ¤ì—”íŠ¸ë¡œí”¼

ë‘ í™•ë¥ ë¶„í¬  ğ‘ ,  ğ‘ ì˜ í¬ë¡œìŠ¤ì—”íŠ¸ë¡œí”¼(cross entropy)  ğ»[ğ‘,ğ‘] ëŠ”

ì´ì‚°í™•ë¥ ë¶„í¬ì˜ ê²½ìš°ì—ëŠ” ë‹¤ìŒê³¼ ê°™ì´ ì •ì˜í•œë‹¤.

$$H[p,q] = -\sum_{k=1}^K p(y_k) \log_2 q(y_k)$$

ì—°ì†í™•ë¥ ë¶„í¬ì˜ ê²½ìš°ì—ëŠ” ë‹¤ìŒì²˜ëŸ¼ ì •ì˜í•œë‹¤.

$$H[p,q] = -\int_{y} p(y) \log_2 q(y) \,dy$$

í¬ë¡œìŠ¤ì—”íŠ¸ë¡œí”¼ëŠ” ì§€ê¸ˆê¹Œì§€ ê³µë¶€í•œ ì—”íŠ¸ë¡œí”¼, ê²°í•©ì—”íŠ¸ë¡œí”¼, ì¡°ê±´ë¶€ì—”íŠ¸ë¡œí”¼ì™€ ë‹¤ë¥´ê²Œ í™•ë¥ ë³€ìˆ˜ê°€ ì•„ë‹Œ í™•ë¥ ë¶„í¬ë¥¼ ì¸ìˆ˜ë¡œ ë°›ëŠ”ë‹¤ëŠ” ì ì— ì£¼ì˜í•˜ë¼.

$$p = [1/4, 1/4, 1/4, 1/4]$$

$$q = [1/2, 1/4, 1/8, 1/8]$$

```python
import numpy as np

-1/4 * np.log2(1/2) - 1/4 * np.log2(1/4) - 1/4 * np.log2(1/8) - 1/4 * np.log2(1/8)
```

### í¬ë¡œìŠ¤ì—”íŠ¸ë¡œí”¼ë¥¼ ì‚¬ìš©í•œ ë¶„ë¥˜ì„±ëŠ¥ ì¸¡ì •

###### í¬ë¡œìŠ¤ì—”íŠ¸ë¡œí”¼ëŠ” ë¶„ë¥˜ëª¨í˜•ì˜ ì„±ëŠ¥ì„ ì¸¡ì •í•˜ëŠ”ë° ì‚¬ìš©ëœë‹¤.

ì´ì§„ë¶„ë¥˜ë¬¸ì œì—ì„œ  ğ‘Œ ëŠ” 0ë˜ëŠ” 1ì´ë‹¤. ë”°ë¼ì„œ  ğ‘ ëŠ”

+ ì •ë‹µì´  ğ‘Œ=1 ì¼ ë•Œ,

$$p(Y=0)=0,  \;  p(Y=1)=1$$

+ ì •ë‹µì´  ğ‘Œ=0 ì¼ ë•Œ

$$p(Y=0)=1,\; p(Y=1)=0$$

ì´ë‹¤.

###### ë¶„í¬  ğ‘ ëŠ”  ğ‘‹ ê°’ì´ ì •í•´ì¡Œì„ ë•Œ ì˜ˆì¸¡ê°’ì˜ í™•ë¥ ë¶„í¬ë‹¤. 

ëª¨ìˆ˜ê°€  ğœ‡ ì¸ ë² ë¥´ëˆ„ì´ë¶„í¬ë¼ê³  ê°€ì •í•œë‹¤.

$$q(Y=0)=1-\mu, q(Y=1)=\mu$$

ë”°ë¼ì„œ í™•ë¥ ë¶„í¬  ğ‘ ì™€  ğ‘ ì˜ í¬ë¡œìŠ¤ì—”íŠ¸ë¡œí”¼ëŠ”

+ ì •ë‹µì´  ğ‘Œ=1 ì¼ ë•Œ,

$$H[p,q] = - {p(Y=0) \log_2 q(Y=0)} - p(Y=1) \log_2 q(Y=1) = 0 -\log_2 \mu$$

+ ì •ë‹µì´  ğ‘Œ=0 ì¼ ë•Œ,

$$H[p,q] = - p(Y=0) \log_2 q(Y=0) - {p(Y=1) \log_2 q(Y=1)} = -\log_2 (1-\mu) - 0$$

ê°€ ëœë‹¤.

###### ì´ ê°’ì€ ë¶„ë¥˜ì„±ëŠ¥ì´ ì¢‹ì„ìˆ˜ë¡ ì‘ì•„ì§€ê³  ë¶„ë¥˜ì„±ëŠ¥ì´ ë‚˜ì ìˆ˜ë¡ ì»¤ì§„ë‹¤. ì´ìœ ëŠ” ë‹¤ìŒê³¼ ê°™ë‹¤.

+ ğ‘Œ=1 ì¼ ë•ŒëŠ”  ğœ‡ ê°€ ì‘ì•„ì§ˆìˆ˜ë¡ í¬ë¡œìŠ¤ì—”íŠ¸ë¡œí”¼ê°€ ì»¤ì§„ë‹¤. ì¦‰, ì˜ˆì¸¡ì´ í‹€ë¦´ìˆ˜ë¡  âˆ’log2ğœ‡ ì˜ ê°’ë„ ì»¤ì§„ë‹¤.

+ ğ‘Œ=0 ì¼ ë•ŒëŠ”  ğœ‡ ê°€ ì»¤ì§ˆìˆ˜ë¡ í¬ë¡œìŠ¤ì—”íŠ¸ë¡œí”¼ê°€ ì»¤ì§„ë‹¤. ì¦‰ ì˜ˆì¸¡ì´ í‹€ë¦´ìˆ˜ë¡  âˆ’log2(1âˆ’ğœ‡) ì˜ ê°’ë„ ì»¤ì§„ë‹¤.

##### ë”°ë¼ì„œ í¬ë¡œìŠ¤ì—”íŠ¸ë¡œí”¼ê°’ì€ ì˜ˆì¸¡ì˜ í‹€ë¦° ì •ë„ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ì˜¤ì°¨í•¨ìˆ˜ì˜ ì—­í• ì„ í•  ìˆ˜ ìˆë‹¤.

ğ‘ ê°œì˜ í•™ìŠµ ë°ì´í„° ì „ì²´ì— ëŒ€í•´ í¬ë¡œìŠ¤ì—”íŠ¸ë¡œí”¼ í‰ê· ì„ êµ¬í•˜ë©´ ë‹¤ìŒ ì‹ìœ¼ë¡œ í‘œí˜„í•  ìˆ˜ ìˆë‹¤. 

###### ì´ ê°’ì„ ë¡œê·¸ì†ì‹¤(log-loss)ì´ë¼ê³ ë„ í•œë‹¤.

$$\text{log loss} = -\dfrac{1}{N} \sum_{i=1}^N \left( y_i \log_2 \mu_i + (1 - y_i) \log_2 (1 - \mu_i) \right)$$

ê°™ì€ ë°©ë²•ìœ¼ë¡œ ì´ì§„ë¶„ë¥˜ê°€ ì•„ë‹Œ ë‹¤ì¤‘ë¶„ë¥˜ì—ì„œë„ í¬ë¡œìŠ¤ì—”íŠ¸ë¡œí”¼ë¥¼ ì˜¤ì°¨ í•¨ìˆ˜ë¡œ ì‚¬ìš©í•  ìˆ˜ ìˆë‹¤. ë‹¤ì¤‘ë¶„ë¥˜ë¬¸ì œì˜ í¬ë¡œìŠ¤ì—”íŠ¸ë¡œí”¼ ì†ì‹¤í•¨ìˆ˜ë¥¼ ì¹´í…Œê³ ë¦¬ ë¡œê·¸ì†ì‹¤(categorical log-loss)ì´ë¼ê³  í•œë‹¤.

$$\text{categorical log loss} = -\dfrac{1}{N} \sum_{i=1}^N \sum_{k=1}^K \left( \mathbb{I}(y_i=k) \log_2 p(y_i=k) \right)$$

ìœ„ ì‹ì—ì„œ  ğ•€(ğ‘¦ğ‘–=ğ‘˜) ëŠ”  ğ‘¦ğ‘– ê°€  ğ‘˜ ì¸ ê²½ìš°ì—ë§Œ 1ì´ê³  ê·¸ë ‡ì§€ ì•Šìœ¼ë©´ 0ì´ ë˜ëŠ” ì§€ì‹œí•¨ìˆ˜(indicator function)ë‹¤.  ğ‘(ğ‘¦ğ‘–=ğ‘˜) ëŠ” ë¶„ë¥˜ëª¨í˜•ì´ ê³„ì‚°í•œ  ğ‘¦ğ‘–=ğ‘˜ ì¼ í™•ë¥ ì´ë‹¤.

ì‚¬ì´í‚·ëŸ°(Scikit-Learn) íŒ¨í‚¤ì§€ì˜ metrics ì„œë¸ŒíŒ¨í‚¤ì§€ëŠ” ë¡œê·¸ì†ì‹¤ì„ ê³„ì‚°í•˜ëŠ” log_loss í•¨ìˆ˜ë¥¼ ì œê³µí•œë‹¤.

ë¶“ê½ƒ ë¶„ë¥˜ë¬¸ì œì—ì„œ ê½ƒë°›ì¹¨ ê¸¸ì´(sepal length) 5.4cmë¥¼ ê¸°ì¤€ìœ¼ë¡œ ì„¸í† ì‚¬ì™€ ë² ë¥´ì‹œì¹¼ë¼ ì¢…ì„ ë¶„ë¥˜í•œ ê²°ê³¼ëŠ” ë‹¤ìŒê³¼ ê°™ë‹¤.

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
idx = np.in1d(iris.target, [0, 1])
X = iris.data[idx, :]
y = iris.target[idx]

df = pd.DataFrame(X, columns=iris.feature_names)
df["y"] = y
df["y_hat"] = (df["sepal length (cm)"] > 5.4).astype(int)  # true 1, false 0
df.tail()
```

ë¡œê·¸ì†ì‹¤ê°’ì„ ê³„ì‚°í•˜ë©´ 3.8ì´ë‹¤.

```python
# yì— ëŒ€í•´ y_hatì„ ê¸°ì¤€ìœ¼ë¡œ ë¶„ë¥˜í•œ ê²°ê³¼ì— ëŒ€í•œ ë¡œê·¸ì†ì‹¤ê°’ êµ¬í•˜ê¸°

from sklearn.metrics import log_loss
log_loss(df["y"], df["y_hat"])
```

(3) ê½ƒë°›ì¹¨ì˜ í­(sepal width)ì— ëŒ€í•´ ìœ„ì˜ ë¶„ì„ì„ ì‹¤ì‹œí•˜ë¼. ì´ ë•ŒëŠ” ê¸°ì¤€ê°’ì´ ì–´ë–»ê²Œ ë˜ëŠ”ê°€?

## ì¿¨ë°±-ë¼ì´ë¸”ëŸ¬ ë°œì‚°

###### ì¿¨ë°±-ë¼ì´ë¸”ëŸ¬ ë°œì‚°(Kullback-Leibler divergence)ì€ ë‘ í™•ë¥ ë¶„í¬  ğ‘(ğ‘¦) ,  ğ‘(ğ‘¦) ì˜ ë¶„í¬ëª¨ì–‘ì´ ì–¼ë§ˆë‚˜ ë‹¤ë¥¸ì§€ë¥¼ ìˆ«ìë¡œ ê³„ì‚°í•œ ê°’ì´ë‹¤.  

ğ¾ğ¿(ğ‘||ğ‘) ë¡œ í‘œê¸°í•œë‹¤.

ì´ì‚°í™•ë¥ ë¶„í¬ì— ëŒ€í•´ì„œëŠ” ë‹¤ìŒê³¼ ê°™ì´ ì •ì˜í•œë‹¤.

\begin{aligned}
KL(p||q)  
&= H[p,q] - H[p] \\
&= \sum_{i=1}^{K} p(y_i) \log_2 \left(\dfrac{p(y_i)}{q(y_i)}\right)
\end{aligned}

ì—°ì†í™•ë¥ ë¶„í¬ì— ëŒ€í•´ì„œëŠ” ë‹¤ìŒì²˜ëŸ¼ ì •ì˜í•œë‹¤.

\begin{aligned}
KL(p || q)  
&= H[p, q] - H[p] \\
&= \int p(y) \log_2 \left(\dfrac{p(y)}{q(y)}\right) dy
\end{aligned}

ì¿¨ë°±-ë¼ì´ë¸”ëŸ¬ ë°œì‚°ì€ í¬ë¡œìŠ¤ì—”íŠ¸ë¡œí”¼ì—ì„œ ê¸°ì¤€ì´ ë˜ëŠ”  ğ‘ ë¶„í¬ì˜ ì—”íŠ¸ë¡œí”¼ ê°’ì„ ëº€ ê°’ì´ë¯€ë¡œ ìƒëŒ€ì—”íŠ¸ë¡œí”¼(relative entropy)ë¼ê³ ë„ í•œë‹¤. 

###### ê·¸ ê°’ì€ í•­ìƒ ì–‘ìˆ˜ì´ë©° ë‘ í™•ë¥ ë¶„í¬  ğ‘(ğ‘¥) ,  ğ‘(ğ‘¥) ê°€ ì™„ì „íˆ ê°™ì„ ê²½ìš°ì—ë§Œ 0ì´ ëœë‹¤.

\begin{aligned}
KL(p || p)  
&= H[p, p] - H[p] \\
&= \int p(y) \log_2 \left(\dfrac{p(y)}{p(y)}\right) dy \\
&= 0
\end{aligned}

ì—­ìœ¼ë¡œ ì¿¨ë°±-ë¼ì´ë¸”ëŸ¬ ë°œì‚°ì´ 0ì´ë©´ ë‘ í™•ë¥ ë¶„í¬ëŠ” ê°™ì€ í™•ë¥ ë¶„í¬ë‹¤. ì¦ëª…ì€ ìƒëµí•œë‹¤.

$$KL(p || q) = 0 \; \leftrightarrow \; p = q$$

ì¿¨ë°±-ë¼ì´ë¸”ëŸ¬ ë°œì‚°ì€ ê±°ë¦¬(distance)ê°€ ì•„ë‹ˆë¼ í™•ë¥ ë¶„í¬  ğ‘ ê°€ ê¸°ì¤€í™•ë¥ ë¶„í¬  ğ‘ ì™€ ì–¼ë§ˆë‚˜ ë‹¤ë¥¸ì§€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ê°’ì´ë¯€ë¡œ ë‘ í™•ë¥ ë¶„í¬ì˜ ìœ„ì¹˜ê°€ ë‹¬ë¼ì§€ë©´ ì¼ë°˜ì ìœ¼ë¡œ ê°’ì´ ë‹¬ë¼ì§„ë‹¤ëŠ” ì ì— ì£¼ì˜í•´ì•¼ í•œë‹¤.

$$KL(p || q) \neq KL(q || p)$$

### ê°€ë³€ê¸¸ì´ ì¸ì½”ë”©ê³¼ ì¿¨ë°±-ë¼ì´ë¸”ëŸ¬ ë°œì‚°

4ê°œì˜ ê¸€ì A, B, C, Dë¡œ ì”Œì—¬ì§„ ë‹¤ìŒê³¼ ê°™ì€ ë¬¸ì„œë¥¼ ê°€ë³€ê¸¸ì´ ì¸ì½”ë”©í•˜ëŠ” ê²½ìš°ë¥¼ ìƒê°í•˜ì.



```python
N = 200
p = [1/2, 1/4, 1/8, 1/8]
doc0 = list("".join([int(N * p[i]) * c for i, c in enumerate("ABCD")]))

np.random.shuffle(doc0)
doc = "".join(doc0)
doc
```

ì´ ë¬¸ì„œë¥¼ êµ¬ì„±í•˜ëŠ” ê¸€ìì˜ í™•ë¥ ë¶„í¬ëŠ” ë‹¤ìŒê³¼ ê°™ë‹¤.

$$p(Y=A)=\dfrac{1}{2},\; p(Y=B)=\dfrac{1}{4},\; p(Y=C)=\dfrac{1}{8},\; p(Y=D)=\dfrac{1}{8}$$

```python
# ìƒì„±í•œ doc ë°ì´í„°ì—ì„œ A, B, C, Dì˜ ë¹„ìœ¨ êµ¬í•˜ê¸°
from collections import Counter

p = np.array(list(Counter(doc).values())) / len(doc)
p
```

í•œ ê¸€ìë‹¹ ì¸ì½”ë”©ëœ ê¸€ììˆ˜ëŠ” ë¶„í¬  ğ‘ ì˜ ì—”íŠ¸ë¡œí”¼ì¸ 1.75ê°€ ëœë‹¤.

\begin{align}
\sum_{k=1}^K p(y_i) \log_2 p(y_i)
&=
- \dfrac{1}{2} \cdot \log_2 \dfrac{1}{2} + 
- \dfrac{1}{4} \cdot \log_2 \dfrac{1}{4} + 
- \dfrac{1}{8} \cdot \log_2 \dfrac{1}{8} + 
- \dfrac{1}{8} \cdot \log_2 \dfrac{1}{8} \\

\dfrac{1}{2} \cdot 1 + 
\dfrac{1}{4} \cdot 2 + 
\dfrac{1}{8} \cdot 3 + 
\dfrac{1}{8} \cdot 3 \\

\end{align}

ì‹¤ì œë¡œ í•œê¸€ìë‹¹ ì¸ì½”ë”©ëœ ê¸€ììˆ˜ëŠ” ë‹¤ìŒê³¼ ê°™ì´ ê³„ì‚°í•  ìˆ˜ ìˆë‹¤.

ì¿¨ë°±-ë¼ì´ë¸”ëŸ¬ ë°œì‚°ì€ ì˜ëª»ëœ ë¶„í¬  ğ‘ ë¡œ ì¸ì½”ë”©ë˜ì—ˆì„ ë•Œì˜ í•œê¸€ìë‹¹ ì¸ì½”ë”©ëœ ê¸€ììˆ˜ì™€ ì›ë˜ì˜ ë¶„í¬  ğ‘ ë¥¼ ì‚¬ìš©í•˜ì˜€ì„ ë•Œ í•œê¸€ìë‹¹ ì¸ì½”ë”©ëœ ê¸€ììˆ˜ì˜ ì°¨ì´ì¸ 0.25ì™€ ê°™ë‹¤.

\begin{aligned}
KL(p||q)  
&= \sum_{i=1}^{K} p(y_i) \log_2 \left(\dfrac{p(y_i)}{q(y_i)}\right) \\
&= - \sum_{i=1}^{K} p(y_i) \log_2 q(y_i) - (- \sum_{i=1}^{K} p(y_i) \log_2 p(y_i)) ) \\
&= H[p, q] - H[p] \\

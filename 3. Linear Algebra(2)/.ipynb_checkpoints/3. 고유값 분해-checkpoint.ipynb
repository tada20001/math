\begin{align}

\end{align}

\begin{align}

\end{align}

\begin{align}
v_j^TAv_j 
= v_j^T \left( \sum_{i=1}^{N} {\lambda_i} v_i v_i^T \right) v_j

> 0
\end{align}

ì´ë¯€ë¡œ ì–‘ìˆ˜ì¸ ê³ ìœ ê°’ë§Œ ê°€ì§„ë‹¤.

\begin{align}

\end{align}

ì´ ì‹ì—ì„œ  ğ‘¥ ê°€  ğ‘£ğ‘– ì™€ ìˆ˜ì§(orthogonal)ì¸ ê²½ìš°ì—ë§Œ 0ì´ ëœë‹¤ëŠ” ê²ƒì„ ì•Œ ìˆ˜ ìˆë‹¤. ê³ ìœ ê°’  ğœ†ğ‘– ê°€ ëª¨ë‘ ì–‘ìˆ˜ì´ë¯€ë¡œ ë”°ë¼ì„œ í–‰ë ¬  ğœ†ğ‘–ğ´ğ‘– ë¥¼ ëª¨ë‘ ë”í•œ í–‰ë ¬  $\lambda_1 A_1 + \cdots + \lambda_N A_N$ ë„ ì–‘ì˜ ì¤€ì •ë¶€í˜¸ë‹¤.

\begin{aligned}

&= \lambda_1 x^TA_1x + \cdots + \lambda_N x^TA_Nx \\
&= \lambda_1 \| x^T v_1 \| ^2 + \cdots + \lambda_N \| x^T v_N \| ^2 \geq 0 \\
\end{aligned}

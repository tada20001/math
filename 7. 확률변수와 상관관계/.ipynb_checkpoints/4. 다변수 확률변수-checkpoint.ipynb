### Í≤∞Ìï© ÌôïÎ•†ÏßàÎüâÌï®Ïàò

Ïù¥Î≤àÏóêÎäî ÌïòÎÇòÏùò Í∞íÏù¥ ÏïÑÎãå Îëê Í∞úÏùò Í∞í, Ï¶â ÌäπÏ†ïÌïú Ïà´Ïûê ÏåçÏù¥ ÎÇòÌÉÄÎÇòÎäî Í≤ΩÏö∞Î•º ÏÉùÍ∞ÅÌïòÏûê. 

Îã®Î≥ÄÏàò Ïù¥ÏÇ∞ÌôïÎ•†Î≥ÄÏàòÏóêÏÑúÏôÄ ÎßàÏ∞¨Í∞ÄÏßÄÎ°ú ÌïòÎÇòÏùò Ïà´Ïûê ÏåçÎßåÏùÑ Í∞ÄÏßÄÎäî ÏÇ¨Í±¥,  Ï¶â ÏïÑÌÜ∞(atom) ÏÇ¨Í±¥Ïóê ÎåÄÌïú ÌôïÎ•†Îßå ÏïåÍ≥† ÏûàÏúºÎ©¥ ÏûÑÏùòÏùò Ïà´Ïûê Ïåç ÏßëÌï© Ï¶â, ÏûÑÏùòÏùò ÏÇ¨Í±¥Ïóê ÎåÄÌï¥ÏÑúÎèÑ ÌôïÎ•†ÏùÑ Í≥ÑÏÇ∞Ìï† Ïàò ÏûàÏúºÎØÄÎ°ú ÌïòÎÇò ÌïòÎÇòÏùò Ïà´Ïûê ÏåçÏóê ÎåÄÌï¥ ÌôïÎ•†ÏùÑ ÏïåÎ†§Ï£ºÎäî ÌôïÎ•†ÏßàÎüâÌï®ÏàòÎßå ÏûàÏúºÎ©¥ Ï†ÑÏ≤¥ ÌôïÎ•†Î∂ÑÌè¨Î•º Ïïå Ïàò ÏûàÎã§. 

Ïù¥Îü¨Ìïú ÌôïÎ•†ÏßàÎüâÌï®ÏàòÎ•º Í≤∞Ìï© ÌôïÎ•†ÏßàÎüâÌï®Ïàò(joint probability mass function)ÎùºÍ≥† ÌïòÎ©∞ Îã§ÏùåÍ≥º Í∞ôÏù¥ ÌëúÏãúÌïúÎã§.

$$p_{XY}(x, y)$$

Ïù¥ ÎïåÎäî ÎÇòÌÉÄÎÇ† Ïàò ÏûàÎäî Ïà´ÏûêÍ∞Ä Îëê Ïà´ÏûêÎ°ú Ïù¥Î£®Ïñ¥ÏßÑ ÏåçÏù¥ÎØÄÎ°ú ÎèÖÎ¶ΩÎ≥ÄÏàòÍ∞Ä  ùë•,ùë¶  Îëê Í∞úÍ∞Ä ÎêúÎã§. Ï¢ÖÏÜçÎ≥ÄÏàòÎäî Í∑∏ Ïà´Ïûê ÏåçÏù¥ ÎÇòÌÉÄÎÇ† ÌôïÎ•†Ïù¥Îã§. ÏòàÎ•º Îì§Ïñ¥ Ïñ¥Îäê ÎåÄÌïôÍµêÏóêÏÑú 50Î™ÖÏùò ÌïôÏÉùÏù¥ X, Y Îëê Í≥ºÎ™©Ïóê ÎåÄÌï¥ ÏãúÌóòÏùÑ Î≥¥Í≥† Í∑∏ Í≤∞Í≥ºÍ∞Ä A, B, C, D, E, F ÌïôÏ†êÏúºÎ°ú ÎÇòÏôîÎã§Í≥† Í∞ÄÏ†ïÌïòÏûê. Í∞Å Í≥ºÎ™©Ïóê ÎåÄÌï¥

### Ï£ºÎ≥Ä ÌôïÎ•†ÏßàÎüâÌï®Ïàò

Ï£ºÎ≥Ä ÌôïÎ•†ÏßàÎüâÌï®Ïàò(marginal probability mass function)Îäî Îëê ÌôïÎ•†Î≥ÄÏàò Ï§ë ÌïòÎÇòÏùò ÌôïÎ•†Î≥ÄÏàò Í∞íÏóê ÎåÄÌï¥ÏÑúÎßå ÌôïÎ•†Î∂ÑÌè¨Î•º ÌëúÏãúÌïú Ìï®ÏàòÏù¥Îã§. Ï¶â Îã§Î≥ÄÏàòÍ∞Ä ÎêòÍ∏∞ Ïù¥Ï†ÑÏùò Îã®Î≥ÄÏàò ÌôïÎ•†ÏßàÎüâÌï®ÏàòÎ•º ÎßêÌïúÎã§.

Í≤∞Ìï© ÌôïÎ•†ÏßàÎüâÌï®ÏàòÏóêÏÑú Ï£ºÎ≥Ä ÌôïÎ•†ÏßàÎüâÌï®ÏàòÎ•º Íµ¨ÌïòÎ†§Î©¥ Ï†ÑÏ≤¥ ÌôïÎ•†Ïùò Î≤ïÏπôÏóê ÏùòÌï¥ Îã§Î•∏ Î≥ÄÏàòÍ∞Ä Í∞ÄÏßà Ïàò ÏûàÎäî Î™®Îì† Í∞íÏùò Í≤∞Ìï©ÌôïÎ•†ÏßàÎüâÌï®ÏàòÎ•º Ìï©Ìïú ÌôïÎ•†Ïù¥ ÎêúÎã§.

\begin{align}%\label{}
\nonumber  p_X(x) = \sum_{y_i} p_{XY}(x,y_i) \\
\nonumber  p_Y(y) = \sum_{x_i} p_{XY}(x_i,y) \\
\end{align}

ÏúÑÏóêÏÑú ÏòàÎ°ú Îì† Ïù¥ÏÇ∞ ÌôïÎ•† Î≥ÄÏàòÏùò Í≤ΩÏö∞Ïóê Í≥ºÎ™© XÎßå Í¥ÄÏã¨Ïù¥ ÏûàÎã§Î©¥ Í≤∞Ìï© ÌôïÎ•†ÏßàÎüâÌï®Ïàò  ùëùùëãùëå(ùë•,ùë¶) Î°úÎ∂ÄÌÑ∞ XÏóê ÎåÄÌïú Ï£ºÎ≥Ä ÌôïÎ•†ÏßàÎüâÌï®Ïàò  ùëùùëã(ùë•) Î•º Íµ¨Ìï¥Ïïº ÌïúÎã§.

Ï£ºÎ≥Ä ÌôïÎ•†ÏßàÎüâÌï®ÏàòÎ•º Í≥ÑÏÇ∞Ìïú Í∞íÏùÄ Îã§ÏùåÍ≥º Í∞ôÎã§.

$$p_X(A) = p_{XY}(A,A) + p_{XY}(A,B) + p_{XY}(A,C) + p_{XY}(A,D) + p_{XY}(A,E) + p_{XY}(A,F) = 0.02$$

```python
pmf_marginal_x = pmf.sum(axis=0)
pmf_marginal_x
```

Ï°∞Í±¥Î∂Ä ÌôïÎ•†ÏßàÎüâÌï®Ïàò(conditional probability mass function)Îäî Îã§Î≥ÄÏàò ÌôïÎ•†Î≥ÄÏàò Ï§ë ÌïòÎÇòÏùò Í∞íÏù¥ ÌäπÏ†ï Í∞íÏúºÎ°ú Í≥†Ï†ïÎêòÏñ¥ ÏÉÅÏàòÍ∞Ä ÎêòÏñ¥ Î≤ÑÎ¶∞ Í≤ΩÏö∞, ÎÇòÎ®∏ÏßÄ Î≥ÄÏàòÏóê ÎåÄÌïú ÌôïÎ•†ÏßàÎüâÌï®ÏàòÎ•º ÎßêÌïúÎã§. ### Îã§Î≥ÄÏàò Ïó∞ÏÜçÌôïÎ•†Î≥ÄÏàò

Ïó∞ÏÜçÌôïÎ•†Î∂ÑÌè¨ÏóêÏÑúÎäî Ïù¥ÏÇ∞ÌôïÎ•†Î∂ÑÌè¨ÏôÄ Í∞ôÏù¥ Îã®ÏàúÏÇ¨Í±¥ÏùÑ Ïù¥Ïö©ÌïòÏó¨ ÌôïÎ•†ÏùÑ Ï†ïÏùòÌï† Ïàò ÏóÜÏúºÎØÄÎ°ú Îã®Î≥ÄÏàò Ïó∞ÏÜçÌôïÎ•†Î≥ÄÏàòÏôÄ Í∞ôÏù¥ ÎàÑÏ†ÅÌôïÎ•†Î∂ÑÌè¨Ìï®ÏàòÎ•º Î®ºÏ†Ä Ï†ïÏùòÌïú ÌõÑ Ïù¥Î•º ÎØ∏Î∂ÑÌïòÏó¨ ÌôïÎ•†Î∞ÄÎèÑÌï®ÏàòÎ•º Ï†ïÏùòÌïòÎäî Î∞©Î≤ïÏùÑ ÏÇ¨Ïö©ÌïúÎã§.

### Í≤∞Ìï© ÎàÑÏ†ÅÌôïÎ•†Î∂ÑÌè¨Ìï®Ïàò

Îëê Ïó∞ÏÜç ÌôïÎ•† Î≥ÄÏàò  ùëã ,  ùëå Ïóê ÎåÄÌïú Í≤∞Ìï© ÎàÑÏ†ÅÌôïÎ•†Î∂ÑÌè¨Ìï®Ïàò  ùëùùëãùëå(ùë•,ùë¶) Îäî Îã§ÏùåÍ≥º Í∞ôÏù¥ Ï†ïÏùòÌïúÎã§.

$$F_{XY}(x, y) = P(\{ X < x \} \cap \{ Y < y \}) = P(\{X < x, Y < y\})$$

Í≤∞Ìï© ÎàÑÏ†ÅÌôïÎ•†Î∂ÑÌè¨Ìï®Ïàò  ùëùùëãùëå(ùë•,ùë¶) Îäî Îã§ÏùåÍ≥º Í∞ôÏùÄ ÌäπÏÑ±ÏùÑ Í∞ÄÏßÑÎã§.

$$F_{XY}(\infty, \infty)=1$$

$$F_{XY}(-\infty, y)=F_{XY}(x,-\infty)=0$$

### Í≤∞Ìï© ÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò

Îã®Î≥ÄÏàò ÌôïÎ•†Î≥ÄÏàòÏùò Í≤ΩÏö∞ÏôÄ Í∞ôÏù¥ Í≤∞Ìï© ÎàÑÏ†ÅÌôïÎ•†Î∂ÑÌè¨Ìï®ÏàòÎ•º ÎØ∏Î∂ÑÌïòÏó¨ Í≤∞Ìï© ÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò(joint probability density function)Î•º Ï†ïÏùòÌï† Ïàò ÏûàÎã§. ÎèÖÎ¶Ω Î≥ÄÏàòÍ∞Ä 2Í∞úÏù¥ÎØÄÎ°ú Í∞ÅÍ∞ÅÏóê ÎåÄÌï¥ Î™®Îëê Ìé∏ÎØ∏Î∂Ñ(partial differentication)Ìï¥Ïïº ÌïúÎã§.

$$p_{XY} = \dfrac{\partial^2 F_{XY}(x, y)}{\partial x \partial y}$$

Í≤∞Ìï© ÌôïÎ•† Î∞ÄÎèÑ Ìï®ÏàòÎ•º ÌäπÏ†ï Íµ¨Í∞ÑÏóê ÎåÄÌï¥ Ï†ÅÎ∂ÑÌïòÎ©¥ Ìï¥Îãπ Íµ¨Í∞ÑÏóê ÎåÄÌïú ÌôïÎ•†Ïù¥ ÎêúÎã§.

$$\int_{x_1}^{x_2} \int_{y_1}^{y_2} p_{XY}(x,y)dxdy = P\big(\{ x_1 \leq X \leq x_2, \; y_1 \leq Y \leq y_2 \}\big)$$

Îî∞ÎùºÏÑú Í≤∞Ìï© ÌôïÎ•† Î∞ÄÎèÑ Ìï®ÏàòÎ•º Î™®Îì† Î≥ÄÏàòÏóê ÎåÄÌï¥  ‚àí‚àû ÏóêÏÑú  ‚àû  ÍπåÏßÄ Ï†ÅÎ∂ÑÌïòÎ©¥ Í∞íÏù¥ 1Ïù¥ ÎêúÎã§.

$$\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} p_{XY}(x,y)dxdy=1$$

Ïó∞ÏÜç ÌôïÎ•† Î≥ÄÏàòÏùò Í≤∞Ìï© ÌôïÎ•† Î∞ÄÎèÑ Ìï®ÏàòÎäî 2Ï∞®Ïõê Ìï®ÏàòÍ∞Ä ÎêúÎã§. 

ÏïÑÎûòÎäî Îã§Î≥ÄÏàò Ï†ïÍ∑ú Î∂ÑÌè¨Ïùò Í≤∞Ìï© ÌôïÎ•† Î∞ÄÎèÑÏùò ÏòàÎ•º Í∑∏Î¶∞ Í≤ÉÏù¥Îã§. Ïñ¥Îñ§ ÏßëÎã®Ïóê ÎåÄÌï¥ XÎäî Î™∏Î¨¥Í≤å, YÎäî ÌÇ§Î•º ÎÇòÌÉÄÎÇ¥Îäî ÌôïÎ•† Î≥ÄÏàòÎùºÍ≥† ÌïòÏûê.

```python
import scipy as sp

mu = [70, 170]
cov = [[150, 140], [140, 300]]  # Í≥µÎ∂ÑÏÇ∞
rv = sp.stats.multivariate_normal(mu, cov)

xx = np.linspace(20, 120, 100)
yy = np.linspace(100, 250, 100)
XX, YY = np.meshgrid(xx, yy)
ZZ = rv.pdf(np.dstack([XX, YY])) # np.dstack 3Ï∞®ÏõêÏúºÎ°ú Î≥ÄÌòï
plt.contour(XX, YY, ZZ)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Joint Probability Density p(x, y)")
plt.show()
```

ÎèôÏùºÌïú Í≤∞Ìï© ÌôïÎ•† Î∞ÄÎèÑ Ìï®ÏàòÎ•º 3Ï∞®ÏõêÏúºÎ°ú Í∑∏Î¶¨Î©¥ ÏïÑÎûòÏôÄ Í∞ôÎã§.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.gca(projection='3d', autoscale_on=True)
ax.contour(XX, YY, ZZ)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xlim(20, 120)
ax.set_ylim(100, 250)
ax.view_init(50, -50)

# Í∑∏Î¶ºÌååÏùº Ï†ÄÏû•
fig.savefig("fig.02.09.01.02.01.png")
plt.close()
from IPython.display import Image
Image("fig.02.09.01.02.01.png")
```

### Ï£ºÎ≥Ä ÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò
Ï£ºÎ≥Ä ÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò(marginal probability density function)Îäî Í≤∞Ìï© ÌôïÎ•†Î∞ÄÎèÑÌï®ÏàòÎ•º ÌäπÏ†ïÌïú ÌïòÎÇòÏùò Î≥ÄÏàòÏóê ÎåÄÌï¥ Í∞ÄÏ§ëÌèâÍ∑†Ìïú Í∞íÏùÑ ÎßêÌïúÎã§. Îî∞ÎùºÏÑú Í≤∞Ìï© ÌôïÎ•†Î∞ÄÎèÑÌï®ÏàòÎ•º ÌïòÎÇòÏùò ÌôïÎ•†Î≥ÄÏàòÏóê ÎåÄÌï¥ÏÑúÎßå Ï†ÅÎ∂ÑÌïòÏó¨ Íµ¨ÌïúÎã§.

Í∞ÄÏ§ëÌèâÍ∑†(Ï†ÅÎ∂Ñ)ÏúºÎ°ú Ïù∏Ìï¥ Ï∞®ÏõêÏù¥ Ìïú Í∞ú Ï§ÑÏñ¥Îì§Í∏∞ ÎïåÎ¨∏Ïóê 2Ï∞®Ïõê ÌôïÎ•† Î≥ÄÏàòÏùò Ï£ºÎ≥Ä ÌôïÎ•† Î∞ÄÎèÑ Ìï®ÏàòÎäî 1Ï∞®Ïõê Ìï®ÏàòÍ∞Ä ÎêúÎã§.

\begin{align}%\label{}
\nonumber  p_X(x) = \int_{-\infty}^{\infty} p_{XY}(x,y)dy \\
\nonumber  p_Y(y) = \int_{-\infty}^{\infty} p_{XY}(x,y)dx
\end{align}

```python
from matplotlib.ticker import NullFormatter
from matplotlib import transforms
from scipy.integrate import simps # Ïã¨Ïä®Î≤ïÏπô(Simpson's rule)ÏùÑ ÏÇ¨Ïö©Ìïú Ï†ÅÎ∂Ñ Í≥ÑÏÇ∞
```

```python
xx = np.linspace(20, 120, 100)
yy = np.linspace(100, 250, 100)
XX, YY = np.meshgrid(xx, yy)
ZZ = rv.pdf(np.dstack([XX, YY]))
fx = [simps(Z, yy) for Z in ZZ.T]
fy = [simps(Z, xx) for Z in ZZ]

plt.figure(figsize=(6, 6))

left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
bottom_h = left_h = left + width + 0.05

rect1 = [left, bottom, width, height]
rect2 = [left, bottom_h, width, 0.2]
rect3 = [left_h, bottom, 0.2, height]

ax1 = plt.axes(rect1)
ax2 = plt.axes(rect2)
ax3 = plt.axes(rect3)

ax2.xaxis.set_major_formatter(NullFormatter())
ax3.xaxis.set_major_formatter(NullFormatter())

ax1.contour(XX, YY, ZZ)
ax2.plot(xx, fx)
base = ax3.transData
rot = transforms.Affine2D().rotate_deg(-90)
plt.plot(-yy, fy, transform=rot + base)

ax1.set_xlim(38, 102)
ax1.set_ylim(120, 220)
ax2.set_xlim(38, 102)
ax3.set_xlim(0, 0.025)
ax3.set_ylim(120, 220)
plt.show()
```

Ï°∞Í±¥Î∂Ä ÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò(conditional probability density function)Îäî Îã§Î≥ÄÏàò ÌôïÎ•† Î≥ÄÏàò Ï§ë ÌïòÎÇòÏùò Í∞íÏù¥ ÌäπÏ†ï Í∞íÏù¥ÎùºÎäî ÏÇ¨Ïã§Ïù¥ ÏïåÎ†§ÏßÑ Í≤ΩÏö∞, Ïù¥Îü¨Ìïú Ï°∞Í±¥(Í∞ÄÏ†ï)Ïóê ÏùòÌï¥ Î≥ÄÌôîÌïú ÎÇòÎ®∏ÏßÄ ÌôïÎ•†Î≥ÄÏàòÏóê ÎåÄÌïú ÌôïÎ•†Î∞ÄÎèÑÌï®ÏàòÎ•º ÎßêÌïúÎã§.

$$p_{X \mid Y}(x \mid y) = \dfrac{p_{XY}(x, y)}{p_{Y}(y)}$$

$$p_{Y \mid X}(y \mid x) = \dfrac{p_{XY}(x, y)}{p_{X}(x)}$$

Ï°∞Í±¥Î∂Ä ÌôïÎ•†Î∞ÄÎèÑÌï®ÏàòÏóêÏÑú Ï°∞Í±¥Ïù¥ ÎêòÎäî ÌôïÎ•†Î≥ÄÏàòÏùò Í∞íÏùÄ ÌäπÏ†ïÌïú Í∞íÏúºÎ°ú Í≥†Ï†ïÎêòÏñ¥ ÏûàÏúºÎØÄÎ°ú Î≥ÄÏàòÍ∞Ä ÏïÑÎãàÎùº Î™®ÏàòÎ°ú ÏÉùÍ∞ÅÌï† Ïàò ÏûàÎã§. ÏòàÎ•º Îì§Ïñ¥  ùëùùëã‚à£ùëå(ùë•‚à£ùë¶) ÏóêÏÑú  ùë¶ Ïùò Í∞íÏùÄ Í≥†Ï†ïÎêòÏñ¥ ÏûàÏúºÎØÄÎ°ú Ïù¥ Í∞íÏùÄ  ùë• Ïùò Ìï®ÏàòÍ∞Ä ÎêúÎã§.

```python
from scipy.integrate import simps # Ï†ÅÎ∂ÑÍ≥ÑÏÇ∞

mag = 10  # ÌôïÎåÄ ÎπÑÏú®

xx = np.linspace(20, 120, 100)
yy = np.linspace(100, 250, 16)
XX, YY = np.meshgrid(xx, yy)
ZZ = rv.pdf(np.dstack([XX, YY]))

plt.figure(figsize=(6, 12))

for i, j in enumerate(range(9, 4, -1)):
    ax = plt.subplot(5, 1, i+1)
    plt.plot(xx, ZZ[j, :] * mag, 'b:', label="joint")
    marginal = simps(ZZ[j, :], xx)
    plt.plot(xx, ZZ[j, :] / marginal, 'r-', label="conditional")
    plt.ylim(0, 0.05)
    ax.xaxis.set_ticklabels([])
    plt.ylabel("p(x, y={:.0f})".format(yy[j]), rotation=0, labelpad=40)
plt.xlabel("x")
plt.tight_layout()
plt.legend()
plt.show()
```

### ÎèÖÎ¶ΩÍ≥º ÏÉÅÍ¥Ä

Îëê ÌôïÎ•†Î≥ÄÏàòÍ∞Ä ÏûàÏùÑ Îïå, Ìïú ÌôïÎ•†Î≥ÄÏàòÏùò ÌëúÎ≥∏ Í∞íÏù¥ Îã¨ÎùºÏßÄÎ©¥ Îã§Î•∏ ÌôïÎ•†Î≥ÄÏàòÏùò Ï°∞Í±¥Î∂Ä Î∂ÑÌè¨Í∞Ä Îã¨ÎùºÏßà Îïå ÏÑúÎ°ú ÏÉÅÍ¥Ä Í¥ÄÍ≥ÑÍ∞Ä ÏûàÎã§Í≥† ÌïúÎã§. Î∞òÎåÄÎ°ú Îëê ÌôïÎ•†Î≥ÄÏàòÍ∞Ä ÏÉÅÍ¥Ä Í¥ÄÍ≥ÑÍ∞Ä ÏïÑÎãàÎ©¥ ÏÑúÎ°ú ÎèÖÎ¶Ω(independent)Ïù¥ÎùºÍ≥† ÌïúÎã§. ÌôïÎ•†Î≥ÄÏàòÏùò ÎèÖÎ¶ΩÏùÑ ÏàòÌïôÏ†ÅÏúºÎ°ú Ï†ïÏùòÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

###### Îëê ÌôïÎ•†Î≥ÄÏàò  ùëã ,  ùëå Ïùò Í≤∞Ìï© ÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò(joint pdf)Í∞Ä Ï£ºÎ≥Ä ÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò(marginal pdf)Ïùò Í≥±Í≥º Í∞ôÏúºÎ©¥ ÏÑúÎ°ú ÎèÖÎ¶Ω(independent)Ïù¥Îã§.

$$p_{XY}(x, y) = p_X(x)p_Y(y)$$

Ïù¥ Ï†ïÏùòÎäî ÌôïÎ•†Î≥ÄÏàòÍ∞Ä Îëê Í∞ú Î≥¥Îã§ ÎßéÏùÑ ÎïåÎèÑ Ï†ÅÏö©ÎêúÎã§. ÏòàÎ•º Îì§Ïñ¥ ÏÑ∏ Í∞úÏùò ÌôïÎ•†Î≥ÄÏàò  ùëã ,  ùëå ,  ùëç Ïùò Í≤∞Ìï© ÌôïÎ•†Î∞ÄÎèÑÌï®ÏàòÍ∞Ä Í∞ÅÍ∞ÅÏùò Ï£ºÎ≥Ä ÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò(marginal pdf)Ïùò Í≥±Í≥º Í∞ôÏúºÎ©¥ ÏÑ∏ ÌôïÎ•†Î≥ÄÏàòÎäî ÏÑúÎ°ú ÎèÖÎ¶ΩÏù¥Îã§.

$$p_{XYZ}(x, y, z) = p_X(x)p_Y(y)p_Z(z)$$

Ïù¥ Îïå  ùëã ,  ùëå ,  ùëç  Ï§ë Ïñ¥Îäê Îëê ÌôïÎ•†Î≥ÄÏàòÎ•º Í≥®ÎùºÎèÑ ÏÑúÎ°ú ÎèÖÎ¶ΩÏù¥ ÎêúÎã§.

\begin{eqnarray}
p_{XY}(x, y) 
&=& \sum_{z\in\Omega_z} p_{XY}(x, y, z) \\
&=& \sum_{z\in\Omega_z}  p_X(x)p_Y(y)p_Z(z) \\
&=& p_X(x)p_Y(y) \sum_{z\in\Omega_z}  p_Z(z) \\
&=&  p_X(x)p_Y(y) 
\end{eqnarray}

### ÎèÖÎ¶ΩÏùò Ïòà: Î∞òÎ≥µÏãúÌñâ

Í∞ôÏùÄ ÌôïÎ•†Î≥ÄÏàòÏóêÏÑú Î≥µÏàòÏùò ÌëúÎ≥∏ Îç∞Ïù¥ÌÑ∞Î•º Ï∑®ÌïòÎäî Í≤ΩÏö∞ÏóêÎäî Ïù¥ ÌëúÎ≥∏Îì§ÏùÄ ÏÑúÎ°ú ÎèÖÎ¶ΩÏù∏ ÌôïÎ•†Î≥ÄÏàòÎì§ÏóêÏÑú ÎÇòÏò® ÌëúÎ≥∏ÏúºÎ°ú Î≥º Ïàò ÏûàÎã§. Îî∞ÎùºÏÑú ÌôïÎ•†Î∞ÄÎèÑÌï®ÏàòÍ∞Ä  ùëì(ùë•) Ïù¥Í≥† ÌëúÎ≥∏ Îç∞Ïù¥ÌÑ∞Í∞Ä  {ùë•1,ùë•2,ùë•3,‚ãØ,ùë•ùëÅ} Ïù¥Î©¥ Ïù¥ Îç∞Ïù¥ÌÑ∞, Ï¶â Î≤°ÌÑ∞  (ùë•1,ùë•2,ùë•3,‚ãØ,ùë•ùëÅ) Í∞Ä ÎÇòÏò¨ ÌôïÎ•†ÏùÄ Îã§ÏùåÍ≥º Í∞ôÎã§.

$$p(x_1, x_2, x_3, \cdots, x_N ) = \prod_{i=1}^N p(x_i)$$

### Ï°∞Í±¥Î∂Ä ÌôïÎ•†Î∂ÑÌè¨

ÎèÖÎ¶ΩÏù∏ Îëê ÌôïÎ•†Î≥ÄÏàò  ùëã,ùëå Ïùò Ï°∞Í±¥Î∂Ä ÌôïÎ•†Î∞ÄÎèÑÌï®ÏàòÎäî Ï£ºÎ≥Ä ÌôïÎ•†Î∞ÄÎèÑÌï®ÏàòÏôÄ Í∞ôÎã§.

$$p_{X \mid Y} (x | y) = \dfrac{p_{XY}(x, y)}{p_{Y}(y)} = \dfrac{p_{X}(x) p_{Y}(y)}{p_{Y}(y)} = p_{X}(x)$$

$$p_{Y \mid X} (y | x) = \dfrac{p_{XY}(x, y)}{p_{X}(x)} = \dfrac{p_{X}(x) p_{Y}(y)}{p_{X}(x)} = p_{Y}(y)$$

ÌôïÎ•†Î≥ÄÏàò  ùëã Í∞Ä Îã§Î•∏ ÌôïÎ•†Î≥ÄÏàò  ùëå Ïóê ÎèÖÎ¶ΩÏù¥Î©¥ Ï°∞Í±¥Î∂Ä ÌôïÎ•† Î∂ÑÌè¨Í∞Ä Ï°∞Í±¥Ïù¥ ÎêòÎäî ÌôïÎ•†Î≥ÄÏàòÏùò Í∞íÏóê ÏòÅÌñ•ÏùÑ Î∞õÏßÄ ÏïäÎäîÎã§. Ï¶â,  ùëå  Í∞íÏù¥  ùë¶1 Ïùº ÎïåÏôÄ  ùë¶2 Ïùº ÎïåÏùò Ï°∞Í±¥Î∂Ä ÌôïÎ•† Î∂ÑÌè¨  ùëì(ùë•‚à£ùë¶1) Í≥º  ùëì(ùë•‚à£ùë¶2) Ïù¥  ùëì(ùë•) Î°ú Í∞ôÎã§Îäî ÏùòÎØ∏Ïù¥Îã§.

ÏòàÎ•º Îì§Ïñ¥ Îã§ÏùåÍ≥º Í∞ôÏùÄ Îëê Ïù¥ÏÇ∞ ÌôïÎ•†Î≥ÄÏàòÏùò Í≤∞Ìï© ÌôïÎ•† Î∂ÑÌè¨Î•º Î≥¥Ïûê.
